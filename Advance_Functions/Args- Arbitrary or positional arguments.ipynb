```python
#*args:-used to pass Non-Keyword Arguments to a  function or Positional Argument 
#A special Operater we can pass to a fucntion that gathers remaining arguments as a Tuple.
#This is just a parameter you can call it whatevery you want as long as its starts with *
#So what it does is,  we have a function defination if wer put *args as a parameter it will collect any 
#number of additional or extra arguments that have been passed in into a single parameter called args or 
#what ever we name this.
#args returns the output as Tuple and **kwargs returns as dictionary (Key value pairs)
```

```python
def sum_all(num1,num2,num3):
    return num1+num2+num3
```

```python
print(sum_all(4,5,6))
```

```python
#Now, what if I call this function in this way
print(sum_all(4,5,6,7,8,9,10))
#It shows an error
```

```python
#To resolve this error, I use, *args
```

```python
def sum_all(*args):
    return sum(args)
```

```python
sum_all(4,5,6,7,8,9,10)
```

```python
#Another example,

def alphabets(a,b,c,*args):
    print(*args) #will print whatever is given after C
```

```python
alphabets('a','b','c','d','e','f','g','h','i','j','k','l','m','n')
```

```python
##args is just a standard name instead of args we can use any name, only point to remember, it should begin with astriex(*)

def nos(*a):
    print(a)

nos(1,2,3,4,5,56,7)   
```

```python
#what if instead of values we are passing a list or tuple as a arguments 

def sum_all(*args):
    total = 0
    for i in args:
        total +=i    #difference between +=1 and =+1 is given below  
    print(total)
```

```python
sum_all([1,2,3])#Results in a error
```

```python
#To rectify this, we need to call with *keyword, below is the example
a=[1,2,3]

sum_all(*a)
```

```python
#Mostly used
i = 0
while i < 10:
        i += 1   #i=i+1
        print(i)
```

```python
def adder(*num):
    sum=0
    for n in num:
        sum=sum+n
    print("sum", sum)
    
adder(3,5)
adder(3,5,6)
adder(5,6,67,7,88)
```
